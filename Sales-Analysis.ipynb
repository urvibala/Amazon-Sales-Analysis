```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('amazon.csv')
```

## Q-1 What are the top 10 most reviewed products?

```python
top_reviewed_products = df.groupby('product_name')['review_id'].count().nlargest(5)
print(top_reviewed_products)
```

## Q-2 What is the distribution of ratings?

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['rating'], bins=5, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

## Q-3 How many products have a discount greater than 50%?

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('[%]', '', regex=True).astype(float)
discount_gt_50 = df[df['discount_percentage'] > 50]['product_id'].count()
print("Number of products with discount greater than 50%:", discount_gt_50)
```

## Q-4 What are the top 5 users who have provided the most reviews?

```python
top_reviewers = df.groupby('user_name')['review_id'].count().nlargest(5)
print(top_reviewers)
```

## Q-5 What are the top 5 most common review titles?

```python
top_revietitles = df.groupby('review_title')['review_id'].count().nlargest(5)
print(top_revietitles)
```

## Q-6 What is the average rating for products with different discount percentages?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='discount_percentage', y='rating', data=df)
plt.title('Average Rating by Discount Percentage')
plt.xlabel('Discount Percentage')
plt.ylabel('Average Rating')
plt.show()
```

## Q-7 What is the distribution of discount percentages?

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['discount_percentage'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
```

## Q-8 How does the average discount percentage vary with user ratings?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='rating', y='discount_percentage', data=df)
plt.title('Average Discount Percentage vs. User Ratings')
plt.xlabel('User Rating')
plt.ylabel('Average Discount Percentage')
plt.show()
```

## Q-9 Which product has the highest number of ratings?

```python
df['rating_count'] = df['rating_count'].str.replace('[,]', '', regex=True).astype(float)
product_with_highest_ratings = df.loc[df['rating_count'].idxmax()]
print("Product with the highest number of ratings:")
print(product_with_highest_ratings[['product_id', 'product_name', 'rating_count']])
```

## Q-10 How many unique users have provided reviews?

```python
unique_users = df['user_id'].nunique()
print("Number of unique users:", unique_users)
```

## Q-11 What is the average discount percentage across all products?

```python
average_discount_percentage = df['discount_percentage'].astype(float).mean()
print("Average Discount Percentage:", average_discount_percentage)
```

## Q-12 How many products have a discount of 0%?

```python
zero_discount_products = df[df['discount_percentage'] == '0%']['product_id'].count()
print("Number of products with 0% discount:", zero_discount_products)
```
